```python
import os, sys
import ogr
import pandas as pd

src = r"F:\Vishal_JRF\RBS\GiZ\Uttarakhand\shps"
fnames = os.listdir(src)
geom1 = []
uid1 = []
name1 = []
flag  = 1
for f in fnames:
    if f[-4:]=='.shp':
        flag = flag + 1
        print (f)
        f1 = os.path.join(src,f)
        shp = ogr.Open(f1)
        lyr = shp.GetLayer()
        #print (lyr.GetSpatialRef())
        featList = range(lyr.GetFeatureCount())
        for FID in featList:
            feature = lyr.GetFeature(FID)
            geom = feature.GetGeometryRef()
            geom1.append(str(geom))
            name= str(feature.GetField("Plot_No"))
            name1.append(name)
            uid = str(feature.GetField("UID"))
            uid1.append(uid)
ds = pd.DataFrame({"Plot_No":name1,"UID":uid1,"Geometry":geom1})
ds
```

```python
dst = r'F:\Vishal_JRF\RBS\GiZ\Uttarakhand\merged_result\\merged_IIS.shp'

spatialref = osr.SpatialReference()  # Set the spatial ref.
spatialref.SetWellKnownGeogCS('WGS84')  # WGS84 aka ESPG:4326

driver = ogr.GetDriverByName("ESRI Shapefile")
dstfile = driver.CreateDataSource(dst) # Your output file

# Please note that it will fail if a file with the same name already exists
dstlayer = dstfile.CreateLayer("layer", spatialref, geom_type=ogr.wkbPolygon) 

# Add the other attribute fields needed with the following schema :
fielddef = ogr.FieldDefn("UID", ogr.OFTString)
fielddef.SetWidth(80)
dstlayer.CreateField(fielddef)

fielddef = ogr.FieldDefn("Plt_No", ogr.OFTString)
fielddef.SetWidth(80)
dstlayer.CreateField(fielddef)
l = 1

for i in range(0,len(geom1)):
    plt = name1[i]
    uid = uid1[i]
    poly = geom1[i]
    poly1 = ogr.CreateGeometryFromWkt(poly)
    feature = ogr.Feature(dstlayer.GetLayerDefn())
    feature.SetGeometry(poly1)
    feature.SetField("UID", uid) # A field with an unique id.
    l = l + 1
    feature.SetField("Plt_No", plt) # And a name (which is in the first field of my test file)
    dstlayer.CreateFeature(feature)
feature.Destroy()
dstfile.Destroy() 
```
